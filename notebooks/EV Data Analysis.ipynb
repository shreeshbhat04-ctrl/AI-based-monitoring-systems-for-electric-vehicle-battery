```python
# Convert the timestamp column to a datetime object
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Sort the entire dataset by timestamp to ensure it's in chronological order
df = df.sort_values(by='timestamp').reset_index(drop=True)

print("Timestamp column converted and dataframe sorted.")
print(df.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

# --- Define the features your model will use ---
# Let's assume you found the correct acceleration column name was 'acceleration'
# If it was different, make sure to use the correct name here.
temporal_features = ['speed_kmh', 'acceleration_mps2']
static_features = ['battery_level', 'traffic_density']
target_feature = 'energy_consumption'

# Combine all features for scaling
all_features = temporal_features + static_features + [target_feature]

# --- Scale the features ---
# Create a new DataFrame with only the features we need
data_to_scale = df[all_features]

# Initialize the scaler
scaler = MinMaxScaler()

# Fit the scaler to the data and transform it
scaled_data = scaler.fit_transform(data_to_scale)

# Convert the scaled data back to a DataFrame for easier use
scaled_df = pd.DataFrame(scaled_data, columns=all_features)

print("Data scaled successfully. Here are the first 5 rows of the scaled data:")
print(scaled_df.head())
```

```python
import numpy as np

def create_sequences(data, sequence_length, temporal_cols, static_cols, target_col):
    """
    Creates sequences for the hybrid LSTM-DNN model.
    """
    X_lstm, X_dnn, y = [], [], []
    
    # Create column indices for faster lookup
    temporal_indices = [data.columns.get_loc(col) for col in temporal_cols]
    static_indices = [data.columns.get_loc(col) for col in static_cols]
    target_index = data.columns.get_loc(target_col)

    for i in range(len(data) - sequence_length):
        # The sequence of temporal data for the LSTM
        lstm_sequence = data.iloc[i:i + sequence_length, temporal_indices].values
        X_lstm.append(lstm_sequence)
        
        # The static data at the end of the sequence for the DNN
        dnn_input = data.iloc[i + sequence_length - 1, static_indices].values
        X_dnn.append(dnn_input)
        
        # The target value we want to predict
        target = data.iloc[i + sequence_length - 1, target_index]
        y.append(target)
        
    return np.array(X_lstm), np.array(X_dnn), np.array(y)

# --- Generate the sequences ---
SEQUENCE_LENGTH = 60 # Look at the last 60 time steps (e.g., 60 seconds)

X_lstm, X_dnn, y = create_sequences(
    scaled_df, 
    SEQUENCE_LENGTH, 
    temporal_features, 
    static_features, 
    target_feature
)

print(f"Sequences created successfully!")
print(f"Shape of LSTM input data (X_lstm): {X_lstm.shape}")
print(f"Shape of DNN input data (X_dnn): {X_dnn.shape}")
print(f"Shape of target data (y): {y.shape}")
```

```python
# Define split sizes
train_size = int(len(X_lstm) * 0.70)
val_size = int(len(X_lstm) * 0.15)
test_size = len(X_lstm) - train_size - val_size

# Split the data
X_lstm_train, X_lstm_val, X_lstm_test = X_lstm[:train_size], X_lstm[train_size:train_size+val_size], X_lstm[train_size+val_size:]
X_dnn_train, X_dnn_val, X_dnn_test = X_dnn[:train_size], X_dnn[train_size:train_size+val_size], X_dnn[train_size+val_size:]
y_train, y_val, y_test = y[:train_size], y[train_size:train_size+val_size], y[train_size+val_size:]

print("--- Data Splitting Complete ---")
print(f"Training set size: {len(X_lstm_train)}")
print(f"Validation set size: {len(X_lstm_val)}")
print(f"Test set size: {len(X_lstm_test)}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, concatenate
from tensorflow.keras.callbacks import EarlyStopping

# --- Define LSTM Input Branch ---
# The shape is (SEQUENCE_LENGTH, num_temporal_features)
lstm_input = Input(shape=(X_lstm_train.shape[1], X_lstm_train.shape[2]), name='lstm_input')
lstm_branch = LSTM(50, activation='relu')(lstm_input)

# --- Define DNN Input Branch ---
# The shape is (num_static_features,)
dnn_input = Input(shape=(X_dnn_train.shape[1],), name='dnn_input')

# --- Concatenate (Combine) the Branches ---
combined_features = concatenate([lstm_branch, dnn_input])

# --- Add Final Dense Layers for Prediction ---
# These layers process the combined information
x = Dense(50, activation='relu')(combined_features)
x = Dense(25, activation='relu')(x)
output = Dense(1, name='output')(x) # Final output neuron for energy_consumption

# --- Create and Compile the Model ---
model = Model(inputs=[lstm_input, dnn_input], outputs=output)

# Compile the model with an optimizer and loss function
model.compile(optimizer='adam', loss='mean_squared_error')

# Print a summary of the model architecture
print("--- Model Architecture ---")
model.summary()
```

```python
# Define the EarlyStopping callback
# It will monitor the validation loss and stop if it doesn't improve for 5 epochs
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# --- Train the model ---
# We provide the training data and the validation data
# Note that the input data is a list: [X_lstm_train, X_dnn_train]
history = model.fit(
    [X_lstm_train, X_dnn_train], 
    y_train,
    epochs=100,
    batch_size=32,
    validation_data=([X_lstm_val, X_dnn_val], y_val),
    callbacks=[early_stopping],
    verbose=1 # This will show a progress bar
)

print("\n--- Model Training Complete ---")
```

```python
# Plot the training and validation loss
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss (Mean Squared Error)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Make predictions on the test data
test_predictions = model.predict([X_lstm_test, X_dnn_test])

# Calculate the error metrics
mse = mean_squared_error(y_test, test_predictions)
mae = mean_absolute_error(y_test, test_predictions)

print("--- Model Evaluation on Test Data ---")
print(f"Mean Squared Error (MSE): {mse:.6f}")
print(f"Mean Absolute Error (MAE): {mae:.6f}")
```

```python
# Plot a sample of the predictions vs the actual values
plt.figure(figsize=(15, 6))
plt.plot(y_test[:200], label='Actual Energy Consumption', color='blue')
plt.plot(test_predictions[:200], label='Predicted Energy Consumption', color='red', linestyle='--')
plt.title('Model Predictions vs. Actual Values (Sample from Test Set)')
plt.xlabel('Time Steps')
plt.ylabel('Scaled Energy Consumption')
plt.legend()
plt.show()
```

```python
# --- Inverse transform the predictions to get real values ---

# Important: The scaler was fitted on multiple columns, so we need to create a dummy array
# with the same shape to perform the inverse transform.
dummy_array = np.zeros((len(test_predictions), len(all_features)))
dummy_array[:, all_features.index(target_feature)] = test_predictions.ravel()

# Now, inverse transform the dummy array
unscaled_predictions = scaler.inverse_transform(dummy_array)

# Extract the unscaled predictions for our target feature
unscaled_predictions = unscaled_predictions[:, all_features.index(target_feature)]

# --- Calculate the average consumption rate ---

# First, get the original, unscaled speed data for the test set
original_test_df = df.iloc[train_size + val_size + SEQUENCE_LENGTH:]

# Average distance traveled per time step (assuming 1 step = 1 second)
# Speed in km/h is converted to km/s
avg_distance_per_step_km = (original_test_df['speed_kmh'].mean()) / 3600 

# Average predicted energy consumption per time step (in original units, likely Wh)
avg_energy_per_step_wh = unscaled_predictions.mean()

# Calculate the consumption rate in Wh per km
consumption_rate_wh_per_km = avg_energy_per_step_wh / avg_distance_per_step_km

print(f"--- Range Calculation Parameters ---")
print(f"Average Predicted Consumption Rate: {consumption_rate_wh_per_km:.2f} Wh/km")
```

```python
def estimate_range(current_soc_percent, total_battery_capacity_kwh, consumption_rate_wh_per_km):
    """
    Estimates the remaining driving range.
    """
    # Convert battery capacity to Wh
    total_battery_capacity_wh = total_battery_capacity_kwh * 1000
    
    # Calculate remaining energy in Wh
    remaining_energy_wh = total_battery_capacity_wh * (current_soc_percent / 100)
    
    # Calculate estimated range
    estimated_range_km = remaining_energy_wh / consumption_rate_wh_per_km
    
    return estimated_range_km

# --- Example Calculation ---
# Let's assume the EV has a 75 kWh battery and the current charge is 65%
my_battery_capacity_kwh = 75
my_current_soc = 65

predicted_range = estimate_range(my_current_soc, my_battery_capacity_kwh, consumption_rate_wh_per_km)

print(f"\n--- Final Range Estimation ---")
print(f"With a battery at {my_current_soc}%, the estimated remaining range is: {predicted_range:.2f} km")
```
